## Exercise 2

Machine Learning pipeline

1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

2. Load the dataset


```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df.drop(columns="Id", inplace=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Species'].value_counts()
```

3. Visualize the dataset

```python
plt.hist(df["SepalLengthCm"])
plt.xlabel("SepalLength")
plt.ylabel("Number of rows")
plt.show()
```

```python
plt.hist(df["PetalLengthCm"])
plt.xlabel("PetalLength")
plt.ylabel("Number of rows")
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(8,6))

ax[0,0].hist(df["SepalLengthCm"])
ax[0,1].hist(df["SepalWidthCm"])
ax[1,0].hist(df["PetalLengthCm"])
ax[1,1].hist(df["PetalWidthCm"])

ax[0,0].set_title("sepal_length")
ax[0,1].set_title("sepal_width")
ax[1,0].set_title("petal_length")
ax[1,1].set_title("petal_width")

plt.show()
```

```python
colors = {'Iris-setosa':"red",
          'Iris-virginica':'blue',
          'Iris-versicolor':'green'}
```

```python
plt.scatter(df["SepalLengthCm"],
            df["SepalWidthCm"],
            c=df['Species'].map(colors))

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
plt.scatter(df["PetalLengthCm"],
            df["PetalWidthCm"],
            c=df['Species'].map(colors))

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
plt.scatter(df["SepalLengthCm"],
            df["PetalLengthCm"],
            c=df['Species'].map(colors))

plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Iris Data Scatter Plot")
plt.show()
```

4. Split the dataset into a training set and a testing set

```python
# Array of features
X = df.drop(columns=['Species']).values

#Array of Labels
y = df['Species'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=5)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)

```

```python
print(X_train)
```

5. Train the model

```python
knn = KNeighborsClassifier(n_neighbors=13)

# Train the model
knn.fit(X_train, y_train)

# Test the model
y_pred = knn.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```
